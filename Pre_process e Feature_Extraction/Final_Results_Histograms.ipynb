```python
import seaborn as sns 
sns.set_context('talk')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
EMG_EDA_ECG_F1_score = pd.read_excel('resultados_modelos_hist.xlsx',sheet_name='EMG_EDA_ECG_F1_score',  usecols=lambda x: 'Unnamed' not in x)
EMG_F1_score = pd.read_excel('resultados_modelos_hist.xlsx',sheet_name='EMG_F1_score',  usecols=lambda x: 'Unnamed' not in x)
ECG_F1_score = pd.read_excel('resultados_modelos_hist.xlsx',sheet_name='ECG_F1_score',  usecols=lambda x: 'Unnamed' not in x)
EDA_F1_score = pd.read_excel('resultados_modelos_hist.xlsx',sheet_name='EDA_F1_score',  usecols=lambda x: 'Unnamed' not in x)

# resultados_modelos_hist.xlsx have 4 sheets related to F1-score values (EMG_EDA_ECG_F1_score, EMG_F1_score, EDA_F1_score and ECG_F1_score)
# EMG_EDA_ECG_F1_score contains the F1 score values obtained by the model evaluation when the features of EMG, EDA and ECG signal are used
# EDA_F1_score contains the F1 score values obtained by the model evaluation when the features of EDA signal are used
# ECG_F1_score contains the F1 score values obtained by the model evaluation when the features of ECG signal are used
```
